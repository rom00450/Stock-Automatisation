```python
import yfinance as yf
import pandas as pd
import openpyxl
from datetime import datetime, timedelta
import os
```

```python
## Define the list of tickers
tickers = [
"LLOY.L",
"BLK",
"BAX",
"COLD",
"CRM",
"DIS",
"EL",
"EZJ.L",
"GAW.L",
"4I7.F",
"IFX.DE",
"MELI",
"NEE",
"PFE",
"TTE.PA",
"SQ",
"^GSPC"]
#MC.PA is LVHM
## Set the end date to today
end_date = datetime.today()
print(end_date)

## Set the interval to collect data
interval = '1wk' 

## Set the start date to 5 years ago
#start_date = end_date - timedelta(days = 5*365)
start_date = datetime(2018,9,24)
print(start_date)
```

```python
## Create an empty DataFrame to store the close prices
close_df = pd.DataFrame()

## Download the close prices for each ticker
for ticker in tickers:
    data = yf.download(ticker, start = start_date, end = end_date, interval = interval)
    close_df[ticker] = data['Close']

## Display the DataFrame
print(close_df)
```

```python
## Set the output folder path
output_folder = r"/Users/roberto/Desktop"

##Reverses the data so the newest data is at the top
reverse_df = close_df.iloc[::-1]

## Export the DataFrame to Excel
if os.path.exists("/Users/roberto/Desktop/SMIF/stock_prices.xlsx"):
    file_path = "/Users/roberto/Desktop/SMIF/stock_prices.xlsx"
    workbook = openpyxl.load_workbook(file_path)
    sheet = workbook['Sheet1']
    reverse_df.to_excel(file_path)
    
else:
    output_file = os.path.join(output_folder, 'stock_prices.xlsx')
    reverse_df.to_excel(output_file)


```
